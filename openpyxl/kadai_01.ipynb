```python
import openpyxl
from datetime import datetime

wb = openpyxl.Workbook()
ws = wb.active

ws.cell(2, 2).value = '請求書'
ws.cell(4, 6).value = 'No.'
ws.cell(4, 7).value = '0001'
ws.cell(5, 6).value = '日付'

cus_info = {
    "会社名": "株式会社ABC", 
    "所在地": "〒101-0022 東京都千代田区神田練塀町300", 
    "TEL": "03-1234-5678",
    "FAX": "03-1234-5678",
    "担当者名": "鈴木一郎"
}

ws.cell(4, 2).value = cus_info["会社名"]
ws.cell(5, 2).value = cus_info["所在地"]
ws.cell(6, 2).value = f'TEL:{cus_info["TEL"]} FAX:{cus_info["FAX"]}'
ws.cell(7, 2).value = f'担当者名:{cus_info["担当者名"]} 様'

ws.cell(10, 2).value = '商品名'
ws.cell(10, 3).value = '数量'
ws.cell(10, 4).value = '単価'
ws.cell(10, 5).value = '金額'

product_name = ["商品A", "商品B"]
product_pcs = [2, 1]
product_price = [10000, 15000]

subtotal = 0

for num2, name in enumerate(product_name):
    row = num2 + 11

    ws.cell(row, 2).value = name
    ws.cell(row, 3).value = product_pcs[num2]
    ws.cell(row, 4).value = product_price[num2]

    amount = product_pcs[num2] * product_price[num2]
    ws.cell(row, 5).value = amount

    subtotal += amount

ws.cell(13, 5).value = subtotal

ws.cell(15, 2).value = '小計'
ws.cell(15, 5).value = subtotal

ws.cell(16, 2).value = '消費税'
ws.cell(16, 5).value = subtotal * 0.1

ws.cell(17, 2).value = '合計'
ws.cell(17, 5).value = ws.cell(15, 5).value + ws.cell(16, 5).value

today = datetime.today()

ws.cell(5, 7).value = today.strftime('%Y/%m/%d')
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

wb.save(filename)
```
