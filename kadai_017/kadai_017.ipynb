```python
class Human:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):

        human_age = self.age
        human_name = self.name

        if human_age >= 20:
            print(f"{human_name}さんは{human_age}歳であり、大人です。")
            
        else:
            print(f"{human_name}さんは{human_age}歳であり、未成年です。大人ではありません。")

human_1 = Human("Tom.K", 30)
human_2 = Human("Nor.K", 18)

human_1.check_adult()
human_2.check_adult()
```

```python
class Human:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        human_name = self.name
        human_age = self.age

        if human_age >= 20:
            print(f"{human_name}さんは{human_age}歳で大人です")

        else:
            print(f"{human_name}さんは{human_age}歳で未成年です。大人ではありません。")

humans = []
humans.append(Human("Tom.K", 30))
humans.append(Human("Nor.K", 19))
humans.append(Human("Fum.K", 43))

for human in humans:
    human.check_adult()
```
