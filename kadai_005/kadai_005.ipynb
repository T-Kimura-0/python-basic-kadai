## 配列
# リスト配列
リストとは、要素の追加・変更・削除が行える配列のこと
- リスト名 = ["要素1", "要素2", "要素3", ・・・・]<br>
※リスト名のインデックス<br>
リスト名 = [0, 1, 2, 3, ．．．]<br>

# 要素の追加、変更、削除
- 追加：リスト名.append(追加する要素の値(数字や文字列))
- 変更：リスト名[インデックス] = 新しい値
- 削除：リスト名.pop(削除する要素のインデックス(数字))

# 要素の取り出し表示
- リスト要素の取り出し表示<br>
print(リスト名[インデックス数])
- リスト全体の取り出し表示<br>
print(リスト名)

```python
user_names = ["侍太郎", "侍一郎", "侍二郎", "侍三郎", "侍四郎"]
user_ages  = [36, 33, 29, 25, 22]

#要素追加
user_names.append("侍五郎")

#要素変更
user_names[1] = "侍花子"
print(user_names)
print("-----")

#要素削除
user_names.pop(2)
print(user_names)
print("-----")

#print()は各個表示されるが、リスト名のみは最後に書かれた文のみが表示される
print(user_names, user_ages)
user_names, user_ages
user_names
print(user_ages)
print(user_names)
print(user_names[2])
```

# タプル配列<br>
タプル(tuple)とは、要素の追加・変更・削除が行えない配列のこと<br>
タプル名 = (要素1, 要素2, 要素3, ・・・)<br>
※リストとカッコの形が違うことに注意
# 要素の追加、変更、削除<br>
タプルは要素の追加、変更、削除はできない。コマンドもない
# 要素の取り出し表示
- タプル要素の取り出し表示<br>
print(タプル名[インデックス数])
- タプル全体の取り出し表示<br>
print(タプル名)


```python
country_names = ("日本", "アメリカ", "イギリス", "フランス" )

# タプルの要素変更はできない
country_names[1] = "ドイツ"
country_names
```

# セット配列
順番、重複の概念を持たない配列（別名：集合）。<br>
セット名 = {要素1, 要素2, 要素3, ・・・}
# 要素の追加、変更、削除
- 追加：セット名.add(追加する要素の値（数値、文字列）)<br>
- 削除：セット名.remove(削除する要素の値（数字、文字列）)
# 要素の取り出し表示
- セット配列全体の取り出し表示<br>
print(セット名)<br>
※セット配列は順番の概念を持たないため、print(セット名[インデックス数])を使って表示できない。

```python
# 1～10までに含まれる素数のセット（2と5が重複）
prime_numbers = {2, 5, 2, 7, 3, 5, 5 }
prime_numbers.add(9)
prime_numbers.remove(2)

country_names = {"日本", "アメリカ", "イギリス", "フランス" ,"日本"}
country_names.add("いんど")
country_names.remove("フランス")

# セット全体を表示
print(prime_numbers,country_names)
print(country_names[3])
```

# 連想配列（ディクショナリ、辞書）
キーと値をワンセットで管理すする配列。インデックスの代わりにキーを指定する。
- 配列の書き方<br>
連想配列名 = {キー1: 値1, キー2: 値2, キー3: 値3, ・・・}<br>
※キーが重複した場合<br>
キーが重複した場合、文の前の値が文の後の値に上書きされる
# 要素の追加、変更、削除
- 追加：連想配列名["新しいキー名"] = "値"
- 変更：連想配列名["変更したいキー名"] = "値"
- 削除：連想配列名.pop("削除したいキー名")
# 要素の取り出し表示
- print(連想配列名["キー名"])
- print(連想配列名)


```python
# 連想配列を作成
personal_data = {"name": "侍太郎", "age": 36, "gender": "男性", "name": "侍花子"}
personal_data["gender"]  = "女性"
personal_data["address"] = "東京都"
personal_data.pop("age")

personal_data
```
